```python
import cv2
import numpy as np

cap = cv2.VideoCapture(0)
def draw_lines(img,lines):
    try:
        for line in lines:
            cords=line[0]
            cv2.line(img,(cords[0],cords[1]),(cords[2],cords[3]),[255,0,0],3)
    except:
        pass
    
    return img
        
while(1):

    _, frame = cap.read()

    cv2.imshow('Original',frame)
    blur=cv2.GaussianBlur(frame,(3,3),0)
    edges = cv2.Canny(blur,100,200)
    lines = cv2.HoughLines(edges,1,np.pi/180,100)
    #print(lines)
    try:
        for rho,theta in lines[0]:
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a*rho
            y0 = b*rho
            x1 = int(x0 + 1000*(-b))   # Here i have used int() instead of rounding the decimal value, so 3.8 --> 3
            y1 = int(y0 + 1000*(a))    # But if you want to round the number, then use np.around() function, then 3.8 --> 4.0
            x2 = int(x0 - 1000*(-b))   # But we need integers, so use int() function after that, ie int(np.around(x))
            y2 = int(y0 - 1000*(a))
            cv2.line(frame,(x1,y1),(x2,y2),(0,255,0),2)
        cv2.imshow('houghlines',frame)
    except:
        pass
    
    k = cv2.waitKey(5) & 0xFF
    if k == 27:
        break

cv2.destroyAllWindows()
cap.release()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
def imshow(img):
    gray =img# cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    plt.imshow(gray)
    plt.title('my picture')
    plt.show()
```

```python
template=cv2.imread('arrow.jpg',cv2.IMREAD_GRAYSCALE)
frame=cv2.imread('dest.png',cv2.IMREAD_GRAYSCALE)
```

```python
temp_canny=cv2.Canny(template,100,200)
imshow(temp_canny)
cv2.imshow('edges',temp_canny)
cv2.waitKey(0)
```

```python
res = cv2.matchTemplate(frame,template,cv2.TM_CCOEFF_NORMED)
threshold = 0.8
loc = np.where( res >= threshold)
w, h = template.shape[::-1]
for pt in zip(loc[::-1]):
    cv2.rectangle(frame, pt, (pt[0] + w, pt[1] + h), (0,255,255), 2)

cv2.imshow('Detected',frame)
cv2.waitKey(0)
```

```python

#cv2.imshow('Original',frame)
blur=cv2.GaussianBlur(frame,(3,3),0)
edges = cv2.Canny(blur,100,200,apertureSize = 3)
#cv2.imshow('edges',edges)
#cv2.waitKey(0)
minLineLength = 5
maxLineGap = 10
lines = cv2.HoughLinesP(edges,1,np.pi/180,minLineLength,minLineLength,maxLineGap)
```

```python
for line in lines:
    print(line)
    x1,y1,x2,y2=line[0]
    cv2.line(edges,(x1,y1),(x2,y2),(255,255,255),3)

cv2.imshow('houghlines',edges)
cv2.waitKey(0)
```
